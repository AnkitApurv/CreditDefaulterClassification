#### Types:
##### Categorical
1. Nominal - Nominal values represent discrete and unordered units with no quantitative values.

2. Ordinal - Ordinal values represent discrete and ordered units.

##### Numerical
1. Discrete - Represents which has no fractional values


2. Continuous - Represents measurements and therefore their values can be fractional.

    a. Interval - Ordered units that have the same difference, don’t have a 'true zero'.
    
    b. Ratio - The same as interval values, but they do have an absolute zero.
